```python
import pandas as pd
from azure.storage.filedatalake import DataLakeServiceClient
from io import StringIO
```

```python
storageAccount = 'Nombre'
sasToken = 'SAS TOKEN DE AZURE'
container = 'nombre'
fileName = 'department.csv'
```

```python
def obtenerClienteADLS():
    client = DataLakeServiceClient(
        accountURL = f'https://{storageAccount}.dfs.core.windows.net',
        credential = sasToken
    )
    return client
```

```python
client = obtenerClienteADLS()
containerClient = client.get_file_system_client(file_system=container)
file = containerClient.get_file_client('fileName')
```

```python
download = file.download_file()
download_bytes = download.readall()
file_content = download_bytes.decode('utf-8')
file_content
```

```python
df = pd.read_csv(StringIO(file_content))
```

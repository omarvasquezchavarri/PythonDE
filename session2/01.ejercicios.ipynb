```python
# funcion para buscar un elemento en una lista
def buscar_elemento(lista, elemento):
    for i in lista:
        if i==elemento:
            return True
    return False
    


print(buscar_elemento(lista, 81))
```

```python
# Crear una función que reciba una lista de palabras y retorne un diccionario con 
# la cantidad de veces que aparece cada palabra.
def contar_palabras(lista):
    dicc = {}
    for palabra in lista:
        if palabra in dicc:
            dicc[palabra] += 1
        else:
            dicc[palabra] = 1
    
    return dicc
    


palabras = ["python", "java", "python", "c++", "python", "java"]
resultado = contar_palabras(palabras)
print(resultado)
```

```python
# Crear una función que reciba un diccionario donde las llaves sean nombres y los valores sean notas.
# Debe retornar el nombre del alumno con la nota más alta
def mejor_alumno1(diccionario):
    nombre = ''
    nota_aux = -1
    for i in diccionario:
        if diccionario[i]>=nota_aux:
            nota_aux = diccionario[i]
            nombre = i
    
    return nombre, diccionario[nombre]

def mejor_alumno2(diccionario):
    nombre = ''
    nota_aux = -1
    for nombreDic, notaDic in diccionario.items():
        if notaDic>=nota_aux:
            nota_aux = notaDic
            nombre = nombreDic
    
    return nombreDic, notaDic

notas = {"Ana": 15, "Luis": 18, "María": 17, "Pedro": 19}
print(mejor_alumno1(notas))  
print(mejor_alumno2(notas))  
```

```python
import pandas as pd
from azure.storage.filedatalake import DataLakeServiceClient
from io import StringIO

storageAccount = 'Nombre'
sasToken = 'SAS TOKEN DE AZURE'
```

```python
#obtener el cliente
def obtenerClienteADLS():
    client = DataLakeServiceClient(
        accountURL = f'https://{storageAccount}.dfs.core.windows.net',
        credential = sasToken
    )
    return client
```

```python
client = obtenerClienteADLS()
```

```python
df = pd.DataFrame({
    'id': [1,2,3],
    'cliente': ['A','B','C'],
    'ventas': [1000,2000,1500]
})
```

```python
container_name = 'silver'
directorio = 'Ventas2025'
file_name = 'ventas.csv'
```

```python
#Si tenemos la certeza que existe el contanedor
containerClient = client.get_file_system_client(file_system=container_name)


try:
    containerClient.create_file_system()
except Exception:
    print(Exception)
    pass
```

```python
dir_client = containerClient.get_directory_client(directorio)
try:
    dir_client.create_directory()
except Exception:
    pass
```

```python
file_client = dir_client.get_file_client(file_name)

csv_buffer = StringIO()

df.to_csv(csv_buffer, index=False)

print(csv_buffer.getvalue())

file_client.upload_data(csv_buffer.getvalue(), overwrite=True)
```

```python
# Nuestra lista de números
numeros = [10, 25, 40, 8, 50, 4]

# Definimos el límite. El programa se detendrá cuando vea un número MENOR a este.
limite = 10

# Empezamos desde el índice 0 (el primer elemento)
i = 0

# El bucle se ejecuta MIENTRAS no hayamos llegado al final de la lista
# Y MIENTRAS el número actual NO sea menor al límite
while i < len(numeros) and numeros[i] >= limite:
    print(numeros[i])
    i += 1  # Avanzamos al siguiente elemento
```

```python
listaEjemplo = [1,5,10,43,20,3,11,4]

numeroVerificiar = 10

i = 0

#print(listaEjemplo)
#print(len(listaEjemplo))
while i < len(listaEjemplo):
    if listaEjemplo[i] >= numeroVerificiar:
        print(listaEjemplo[i])
    #print(i)
    i += 1
```

```python
listaEjemplo = [1,5,10,43,20,3,11,4]

for item in listaEjemplo:
    if item>=numeroVerificiar:
        print(item)
```

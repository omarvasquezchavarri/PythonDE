```python
import pandas as pd
```

```python
path = '../session2/data_retail'

path_customers = f'{path}/customers'
path_orders = f'{path}/orders'
path_order_items = f'{path}/order_items'
path_products = f'{path}/products'
path_categories = f'{path}/categories'
path_departments = f'{path}/departments'
```

```python
df_customers = pd.read_csv(path_customers,sep='|', header=None)
df_orders = pd.read_csv(path_orders,sep='|', header=None)
df_order_items = pd.read_csv(path_order_items,sep='|', header=None)
df_products = pd.read_csv(path_products,sep='|', header=None)
df_categories = pd.read_csv(path_categories,sep='|', header=None, names=['id_categories','id_grupo_categories','desc_categories'])
df_departments = pd.read_csv(path_departments,sep='|', header=None)
```

```python
df_departments.columns =['id_department','desc_deparment']
```

```python
def validar_department(df, columaName):
    if df[columaName].duplicated().any():
        print(f'Hay departamentos duplicados en el dataframe en la columna {columaName}')
        exit(1)
    else:
        print(f'No existen duplicados')

#validar_department(df_departments,'desc_deparment')
validar_department(df_categories,'id_grupo_categories')
```
